```python
import re

def extract_ticket_number(ticket):
    match = re.findall(r'\d+', str(ticket))
    if match:
        return int(match[0])
    else:
        return None

train_data['Ticket_Number'] = train_data['Ticket'].apply(extract_ticket_number)
```

```python
survived_with_ticket_5 = train_data[(train_data['Survived'] == 0) & (train_data['Ticket_Number'] == 5.0)]

count_survived_with_ticket_5 = survived_with_ticket_5.shape[0]

print("Count of individuals who survived with Ticket_Number = 5:", count_survived_with_ticket_5)
```

```python
from sklearn.ensemble import RandomForestRegressor

age_features = ['Pclass', 'SibSp', 'Parch', 'Fare', 'Sex_female', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S']

known_age_data = encoded_data[encoded_data['Age'].notnull()]
unknown_age_data = encoded_data[encoded_data['Age'].isnull()]

rf_age_model = RandomForestRegressor()
rf_age_model.fit(known_age_data[age_features], known_age_data['Age'])

predicted_ages = rf_age_model.predict(unknown_age_data[age_features])

encoded_data.loc[encoded_data['Age'].isnull(), 'Age'] = predicted_ages
```
