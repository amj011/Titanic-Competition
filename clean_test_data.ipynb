```python
import re

def extract_ticket_number(ticket):
    match = re.findall(r'\d+', str(ticket))
    if match:
        return int(match[0])
    else:
        return None

train_data['Ticket_Number'] = train_data['Ticket'].apply(extract_ticket_number)
```
